```python
#Now, x_train,x_test, and x_val must be divided by 255.0 for normalization.
print(len(x_train))
print(len(x_test))
print(len(x_val))
train_x=np.array(x_train)
test_x=np.array(x_test)
val_x=np.array(x_val)

train_x = np.reshape(train_x, (len(train_x), 224, 224, 3))
test_x = np.reshape(test_x, (len(test_x), 224, 224, 3))
val_x = np.reshape(val_x, (len(val_x), 224, 224, 3))
```

```python
train_x=train_x/255.0
test_x=test_x/255.0
val_x=val_x/255.0
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)
val_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(train_path,
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'sparse')
test_set = test_datagen.flow_from_directory(test_path,
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'sparse')
val_set = val_datagen.flow_from_directory(val_path,
                                            target_size = (224, 224),
                                            batch_size = 32,
                                            class_mode = 'sparse')
```

```python
train_y=training_set.classes
test_y=test_set.classes
val_y=val_set.classes
```

```python
training_set.class_indices
```

```python
train_y.shape,test_y.shape,val_y.shape
```

```python
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.applications.vgg19 import preprocess_input
IMAGE_SIZE = [224, 224] #Default image size for VGG16
vgg = VGG19(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
#do not train the pre-trained layers of VGG-19
for layer in vgg.layers:
    layer.trainable = False
```

```python
x = Flatten()(vgg.output)
#adding output layer.Softmax classifier is used as it is multi-class classification
prediction = Dense(6, activation='softmax')(x)

model = Model(inputs=vgg.input, outputs=prediction)
```

```python
# view the structure of the model
model.summary()
```

```python
model.compile(
  loss='sparse_categorical_crossentropy',
  optimizer="adam",
  metrics=['accuracy']
)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=2,patience=5)
#Early stopping to avoid overfitting of model
```

```python
#print(training_set[0])
# fit the model
#history = model.fit(train_x, train_y, validation_data=(val_x,val_y), epochs=10, callbacks=[early_stop],  batch_size=32,shuffle=True)
from tensorflow.keras.optimizers import Adam

#history = model.fit(train_x, train_y, validation_data=(val_x,val_y), epochs=10, batch_size=32)
# Define the optimizer (you can choose a different one if needed)
optimizer = Adam(learning_rate=0.001)

# Compile the model
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Use fit_generator instead of fit
#history = model.fit_generator(training_set, validation_data=val_set, epochs=10)
history = model.fit(training_set, validation_data=val_set, epochs=10)
```

```python
# accuracies
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend()
plt.savefig('vgg-acc-rps-1.png')
plt.show()
```

```python
# loss
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.savefig('vgg-loss-rps-1.png')
plt.show()
```

```python
model.evaluate(test_x,test_y,batch_size=32)
```
